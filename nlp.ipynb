```python
# Importar el pipeline desde transformers
from transformers import pipeline
```

```python
# Cargar el pipeline de NLI con el modelo roberta-large-mnli
nli_pipeline = pipeline("text-classification", model="roberta-large-mnli")
```

```python
classifier = pipeline('zero-shot-classification', model='roberta-large-mnli')
```

```python
sequence_to_classify = "Meal was below average from LHR to Mia"
candidate_labels = ['seat confort', 'cabin staff service', 'food and beverages']
resultados=classifier(sequence_to_classify, candidate_labels)
```

```python
def clasificacion(texto):
    sequence_to_classify = texto
    candidate_labels = ['seat confort', 'cabin staff service', 'food and beverages']
    resultados=classifier(sequence_to_classify, candidate_labels)
    return resultados['labels'][np.argmax(resultados['scores'])]
```
